```python
f = open("dataset2.csv")
Xtest = []
Xtrain = []
Ytest = []
Ytrain = [] 
c=0
for line in f : 
    if c % 3 == 0:
        t = line.split(",")
        Xtrain.append(t[1:])
        Ytrain.append(t[0])
        
    else: 
        t = line.split(",")
        Xtest.append(t[1:]) 
        Ytest.append(t[0])
    c+=1

print(Xtrain)
print(Ytrain)

Xtrain_normalized = preprocessing.normalize(Xtrain, norm='l2')

Xtest_normalized = preprocessing.normalize(Xtest, norm='l2')
```

```python
clf = make_pipeline(StandardScaler(), SVC(gamma='auto'))
clf.fit(Xtrain, Ytrain)
```

```python
clf.score(Xtest,Ytest)
```
